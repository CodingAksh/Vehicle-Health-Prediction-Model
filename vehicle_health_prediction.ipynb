```python
import pandas as pd
import kagglehub
import os
```

```python
# Download latest version
path = kagglehub.dataset_download("parvmodi/automotive-vehicles-engine-health-dataset")

print("Path to dataset files:", path)
```

```python
for filename in os.listdir(path):
    if filename.endswith(".csv"):
        filepath = os.path.join(path, filename)
        break
```

```python
dataset = pd.read_csv(filepath)
```

```python
dataset.head()
dataset.describe()
```

```python
dataset.rename(columns={'Engine rpm': 'engine_rpm', 'Lub oil pressure': 'lub_oil_pressure',
        'Fuel pressure': 'fuel_pressure', 'Collant pressure': 'collant_pressure',
        'lub oil temp': 'lub_oil_temp', 'Collant temp': 'collant_temp'},inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = dataset.drop('Engine Condition', axis=1)
y = dataset['Engine Condition']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)

accuracy = accuracy_score(Y_test, y_pred)
print(f"Accuracy: {accuracy}")
report = classification_report(Y_test, y_pred)
print(f"Classification Report:\n{report}")
```

```python
import numpy as np
# Sample custom data for testing
custom_data = np.array([[700,	2.493592,	11.790927,	3.178981,	84.144163,	81.632187	]])  # Example values
```

```python
# Assuming 'model' is your trained model and 'custom_data' is your input
y_pred_custom = model.predict(custom_data)
confidence_custom = model.predict_proba(custom_data)[:,1]  # Get the confidence score

# Print results
print("Predicted Condition:", y_pred_custom[0])
print("Confidence Level:", confidence_custom[0])
```
